```python
import pandas as pd
from pandas import DataFrame
import random
```

```python

#aluminium	arsenic	cadmium	chloramine	chromium

# Create empty lists to store data for later
alu_list = []
arse_list = []
cad_list = []
chlora_list = []
chromium_list = []
is_safe_list = []

# Create 6000 rows of data
for _ in range(6000):
    # aluminium generate values 
    alu = random.uniform(0.1, 2.8) #range of values specified 
    alu2 = float("{:.2f}".format(alu))
    # arsenic
    arse = random.uniform(0.001, 0.02)
    arse2 = float("{:.3f}".format(arse))
    #cadmium
    cad = random.uniform(0.001, 0.005)
    cad2 = float("{:.3f}".format(cad))
    #chloramine
    chlora = random.uniform(0.1, 4.0)
    chlora2 = float("{:.2f}".format(chlora))
    #chromium
    chromium = random.uniform(0.01, 0.1)
    chromium2 = float("{:.2f}".format(chromium))
    for y in range(30):
        is_safe = 1

# append data to empty list created
    alu_list.append(alu2)
    arse_list.append(arse2)
    cad_list.append(cad2)
    chlora_list.append(chlora2)
    chromium_list.append(chromium2)
    is_safe_list.append(is_safe)
```

```python
# store data to a dictionary
data = {
        "aluminium": alu_list,
        "arsenic":arse_list,
        "cadmium":cad_list,
        "chloramine":chlora_list,
        "chromium":chromium_list,
        "is_safe": is_safe_list
    }

#checking is all data same number of rows
for val in data.values():
    print(len(val))
    
```

```python
# put dictionary to a dataframe
df = DataFrame(data, columns=[
        "aluminium",
        "arsenic",
        "cadmium",
        "chloramine",
        "chromium",
        "is_safe",
    ])
```

```python
# write to new csv file
df.to_csv('/content/drive/MyDrive/Degree/Intro to Data Science/Assignment Test/water_pred.csv')
```
